## Transformação de variaveis catoricas em numericas

```python
previsores = DB.iloc[:,0:14].values # Selecionando atributos previsores - Todas as linhas das colunas "age" até "native-country"
pd.DataFrame(previsores)
```

```python
classe = DB.iloc[:,14] #Selecionando atributos classificadores - Todas as linhas da coluna "income"
pd.DataFrame(classe)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
labelencoder_previsores = LabelEncoder()
```

```python
previsores[:,1] = labelencoder_previsores.fit_transform(previsores[:,1]) # Transformando os dados da coluna "workclass" de catoricos para numericos
```

```python
pd.DataFrame(previsores)
```

```python
# Transformando os dados das colunas de catoricos dos previsores para numericos
previsores[:,3] = labelencoder_previsores.fit_transform(previsores[:,3])
previsores[:,5] = labelencoder_previsores.fit_transform(previsores[:,5])
previsores[:,6] = labelencoder_previsores.fit_transform(previsores[:,6])
previsores[:,7] = labelencoder_previsores.fit_transform(previsores[:,7])
previsores[:,8] = labelencoder_previsores.fit_transform(previsores[:,8])
previsores[:,9] = labelencoder_previsores.fit_transform(previsores[:,9])
previsores[:,13] = labelencoder_previsores.fit_transform(previsores[:,13])
```

```python
pd.DataFrame(previsores)
```

```python
#Transformando  as colunas categoricas com variveis "Dummy" - para os previsores
onehotencoder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [1,3,5,6,7,8,9,13])],remainder='passthrough')
```

```python
previsores = onehotencoder.fit_transform(previsores).toarray()
```

```python
pd.DataFrame(previsores)
```

```python
#Transformando as colunas categoricas dos classificadores para numericos
labelencoder_classe = LabelEncoder()
```

```python
classe = labelencoder_classe.fit_transform(classe)
```

```python
pd.DataFrame(classe)
```

## Escalonamento dos atributos

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
previsores = scaler.fit_transform(previsores)
```

```python
pd.DataFrame(previsores)
```

## Divisão de base de dados em treinamento e teste

```python
# Utilizaremos 25% da base de dados para teste (test_size=0.25)

from sklearn.model_selection import train_test_split
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25, random_state=0)
```

```python
pd.DataFrame(previsores_treinamento)
```

```python
pd.DataFrame(previsores_teste)
```

```python
pd.DataFrame(classe_treinamento)
```

```python
pd.DataFrame(classe_teste)
```
